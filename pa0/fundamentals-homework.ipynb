## Problem 3

Create a 5x5 matrix with random values between -1 and 1, then clamp all values below 0 to 0. Try making this as complicated or simple as you want.

```python
## Code here
this_mat= np.random.uniform(low= -1 , high = 1, size=(5,5))
print(this_mat)
```

```python
# now to clamp the values below 0 to 0 using a naive for loop
def clamper(mat):
    mat_shape = mat.shape
    for i in range(mat_shape[0]):
        for j in range(mat_shape[1]):
            if (mat[i][j] < 0):
                mat[i][j] = 0
    return mat
naive_way = clamper(this_mat)
print(naive_way)


# can also do this using list comprehension
sample_mat = this_mat
comprehension_way = [[ 0 if i < 0 else i for i in sample_mat_row] for sample_mat_row in sample_mat]
print(comprehension_way)
```

```python
# doing it in place without making new matrix using the clip function
np.clip(this_mat , 0, None)
print(this_mat)
```
